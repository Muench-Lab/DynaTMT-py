```python
# Merge the datasets on the 'Accession' column using an outer join
colname_x = 'Abundance: F2: 126, Sample'
colname_y = 'Abundances (Normalized): F2: 126, Sample'

merged_data_foldchange = pd.merge(sumNorm[['Master Protein Accessions', colname_x]], 
                       peptides[['Master Protein Accessions', colname_y]], 
                       on='Master Protein Accessions', 
                       how='inner')

# Replace NaN values with 0
# merged_data.fillna(0, inplace=True)
# Replace NaN values with a small positive number to avoid log(0)
merged_data_foldchange.fillna(0, inplace=True)  # 0.1 is arbitrary; adjust as needed

merged_data_protein = (
    merged_data_foldchange.groupby(['Master Protein Accessions'])
    .agg('sum')
    .reset_index()
)

# Define the data for the plot
x = np.log2(merged_data_protein[colname_x])
y = np.log2(merged_data_protein[colname_y])

# Create the plot
fig, ax = plt.subplots(figsize=(10, 10))

# # Scatter plot
sns.scatterplot(x=x, y=y, ax=ax, s=8)

# 45-degree line
ax.plot([min(x), max(y)], [min(x), max(y)], color='red', linewidth=2)

# Calculate R and R^2
correlation_matrix = np.corrcoef(x, y)
r = correlation_matrix[0, 1]
r_squared = r**2

print(r, r_squared)

# Display the counts, percentages, and R^2 on the plot
text_str = (
    f'R = {r:.4f}'
)
ax.text(0.05, 0.95, text_str, transform=ax.transAxes, fontsize=12, verticalalignment='top', bbox=dict(boxstyle='round,pad=0.5', facecolor='white', alpha=0.5))
ax.set_xlabel(colname_x)
ax.set_ylabel(colname_y)
ax.set_title(f'Correlation between {colname_x} and {colname_y}')

# Set axes to be equal, so the plot is square
ax.axis('equal')

# Save the figure BEFORE plt.show()
plt.savefig('correlation_normalization_sumintenstiy_vs_PD_norm_protein_level.png')

plt.show()
```

```python
def aggregate_psm_data(psm_df):
    """Aggregate PSMs to the peptide level."""
    # Grouping by Annotated Sequence and Modifications and aggregating other data
    aggregation_functions = {
        'Master Protein Accessions': lambda x: '; '.join(set(x.dropna().astype(str))),  # Convert to string and drop NaNs
        # Example for abundance columns, summing up the values
        'Abundance: 126': 'sum',
        'Abundance: 127N': 'sum',
        'Abundance: 127C': 'sum',
        'Abundance: 128N': 'sum',
        'Abundance: 128C': 'sum',
        'Abundance: 129N': 'sum',
        'Abundance: 129C': 'sum',
        'Abundance: 130N': 'sum',
        'Abundance: 130C': 'sum',
        'Abundance: 131N': 'sum',
        # Include other abundance columns as needed
    }
    grouped_df = psm_df.groupby(['Annotated Sequence', 'Modifications']).agg(aggregation_functions).reset_index()
    return grouped_df
# Aggregate the PSM data
aggregated_data = aggregate_psm_data(psms)
```

```python
import re

def standardize_modifications(mod_str):
    # Replace 'N-Term(TMT6plex)' with '1xTMT6plex [N-Term]'
    mod_str = re.sub(r'N-Term\(([^)]+)\)', r'1x\1 [N-Term]', mod_str)

    # Replace other modifications like 'K6(TMT6plex)' with '1xTMT6plex [K6]'
    mod_str = re.sub(r'([KRQ][0-9]+)\(([^)]+)\)', r'1x\2 [\1]', mod_str)

    # Handle specific cases like 'R8(Label:13C(6)15N(4))'
    mod_str = re.sub(r'([KRQ][0-9]+)\(Label:13C\(6\)15N\(4\)\)', r'1xLabel:13C(6)15N(4) [\1]', mod_str)

    # Extract modifications
    mods = re.findall(r'1x[^;]+', mod_str)

    # Sort modifications
    sorted_mods = sorted(mods)

    # Rebuild the modification string
    standardized_mod_str = '; '.join(sorted_mods)
    # Add more patterns or adjust existing ones based on your data
    
    return standardized_mod_str
    
aggregated_data['Modifications'] = aggregated_data['Modifications'].apply(standardize_modifications)
```
