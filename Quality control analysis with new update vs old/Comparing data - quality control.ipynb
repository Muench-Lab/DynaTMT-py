```python
import pandas as pd
```

```python
# data_PSM = pd.read_excel('Result_LMM_2023.xlsx', sheet_name=None)
# data_PSM.keys() # list sheet names # dict_keys(['Info', 'Results'])
data_PSM = pd.read_excel('Result_LMM_2023.xlsx', sheet_name="Results")
# print(data_PSM.columns)
# data_PSM.head()
data_PSM.info()
```

```python
channels = [col for col in data_PSM.columns if ('DMSO' in col or 'CCCP' in col) and '/' not in col]
correlation_PSM = data_PSM[channels].corr()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,10))  # Optional: Sets the figure size
sns.heatmap(correlation_PSM, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=.5)
plt.show()
# Save the plot to a file
plt.savefig('correlation_matrix_PSM.png')
```
