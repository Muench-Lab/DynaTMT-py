```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming data_PSM and data_PSM_boost are your datasets
# Calculate the average of 'DMSO', 'DMSO.1', and 'DMSO.2' for both datasets
data_PSM['Average_DMSO_PSM'] = data_PSM[['DMSO', 'DMSO.1', 'DMSO.2']].mean(axis=1)
data_PSM_boost['Average_DMSO_PSM_boost'] = data_PSM_boost[['DMSO', 'DMSO.1', 'DMSO.2']].mean(axis=1)

# Merge the datasets on the 'Accession' column using an outer join
merged_data = pd.merge(data_PSM[['Accession', 'Average_DMSO_PSM']], 
                       data_PSM_boost[['Accession', 'Average_DMSO_PSM_boost']], 
                       on='Accession', 
                       how='outer')

# Replace NaN values with 0
# merged_data.fillna(0, inplace=True)
# Replace NaN values with a small positive number to avoid log(0)
merged_data.fillna(0.1, inplace=True)  # 0.1 is arbitrary; adjust as needed
```

```python
# Define the data for the plot
# x = merged_data['Average_DMSO_PSM']
# y = merged_data['Average_DMSO_PSM_boost']

# Log-transform the data
x = np.log2(merged_data['Average_DMSO_PSM'])
y = np.log2(merged_data['Average_DMSO_PSM_boost'])


# Create the plot
plt.figure(figsize=(10, 10+))

# Scatter plot
sns.scatterplot(x=x, y=y)

# Regression line
sns.regplot(x=x, y=y, scatter=False, color='red')

# Get R and R² values
correlation_matrix = np.corrcoef(x, y)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
r_value = correlation_xy

# Display R and R² on the plot
plt.text(min(x), max(y), f'R = {r_value:.2f}\nR² = {r_squared:.2f}', fontsize=12, ha='left', va='top', color='red')

plt.xlabel('Average_DMSO_PSM')
plt.ylabel('Average_DMSO_PSM_boost')
plt.title('Correlation between Average_DMSO_PSM and Average_DMSO_PSM_boost')
plt.show()
```

```python
# Create the plot
fig, ax = plt.subplots(figsize=(10, 10))

# Scatter plot
sns.scatterplot(x=x, y=y, ax=ax)

# Regression line
sns.regplot(x=x, y=y, scatter=False, color='red', ax=ax, ci=None)  # Set ci=None to remove the default confidence interval

# Get regression line values
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
regression_line = slope * x + intercept

# Define the standard deviation lines at 0.5 and -0.5 from the regression line
upper_std = regression_line + 0.5
lower_std = regression_line - 0.5

# Plot standard deviation lines
plt.fill_between(x, upper_std, lower_std, color='red', alpha=0.2)  # Adjust alpha for desired transparency

# Display R and R² on the plot
ax.text(min(x), max(y), f'R = {r_value:.2f}\nR² = {r_squared:.2f}', fontsize=12, ha='left', va='top', color='red')

ax.set_xlabel('log2(CCCP/DMSO)_PSM')
ax.set_ylabel('log2(CCCP/DMSO)_boost')
ax.set_title('Correlation between log2(CCCP/DMSO)_PSM and log2(CCCP/DMSO)_PSM_boost')

# Set axes to be equal, so the plot is square
ax.axis('equal')

plt.show()
```
